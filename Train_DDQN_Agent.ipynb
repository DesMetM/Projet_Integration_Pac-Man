```python
from IA.agent_ddqn import AgentDDQN
import os
from IA.environnement import PacEnv
from modele.jeu import Jeu
import pygame
```

```python
pygame.font.init()
env = PacEnv(Jeu(30))
```

```python
agent = AgentDDQN(len(env.observation_space[0][0]),4)
agent.epsilon = 1.0
agent.epsilon_min = 0.05
agent.gamma = 0.95
```

```python
name = "new_reward" + "-inator"
output_dir = "ressource/IA/" + name
```

```python
loaded_weights = 0
#agent.epsilon = agent.epsilon_min
agent.load(output_dir + "/" + name+ "_{:07d}.hdf5".format(loaded_weights))
```

```python
n_episodes = 1000001
batch_size = 32
```

```python
state = env.reset()
```

```python
done = False
frame = 0

for e in range(loaded_weights, n_episodes):
    state = env.reset()
    position = (0, 0) # Il faut compter le nombre de frame o est-ce que Pac-Man est sur la mme tuile.
    
    for time in range(2000):
        
        action = agent.act(state)
        
        next_state, reward, done, info = env.step(action, fantome=True)
        
        if position != info[1] or frame == 4:
            agent.remember(state, action, reward, next_state, done)
            frame = 0
        else:
            frame += 1
        
        
        while not info[0] and not done:
            next_state, reward, done, info = env.step(4)
            
        state = next_state

        
        if done:
            break
            
    print("episode: {}/{}, score: {}, epsilon:{:.2}".format(e,n_episodes,env.last_score, agent.epsilon))
       
    if len(agent.memory) > batch_size:
        agent.replay(batch_size)
    
    
    if e % 1000 == 0 and e != loaded_weights:
        agent.save(output_dir+ "/" + name + "_{:07d}.hdf5".format(e+loaded_weights))
    
        
#agent.save(output_dir+ "/" + name + "_{:05d}.hdf5".format(n_episodes-1))
#agent.save(output_dir+ "/" + name + "_02000.hdf5")
```
