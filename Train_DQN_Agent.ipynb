```python
from IA.agent_dqn import AgentDQN
import os
from IA.environnement import PacEnv
from modele.jeu import Jeu
import pygame
```

```python
pygame.font.init()
env = PacEnv(Jeu(30))
```

```python
agent = AgentDQN(len(env.observation_space[0][0]),4)
agent.epsilon = 1.0
agent.gamma = 0.95
```

```python
name = "trimmed-inator"
output_dir = "ressource/IA/" + name
```

```python
agent.load(output_dir + "/" + name+ "_00000.hdf5")
```

```python
n_episodes = 50001
batch_size = 32
```

```python
state = env.reset()
agent.epsilon = 0.01
#agent.gamma = 0.8
```

```python
done = False
frame = 0

for e in range(n_episodes):
    state = env.reset()
    position = (0, 0) # Il faut compter le nombre de frame où est-ce que Pac-Man est sur la même tuile.
    
    for time in range(5000):
        
        action = agent.act(state)
        
        next_state, reward, done, info = env.step(action)
        
        if position != info[1] or frame == 4:
            agent.remember(state, action, reward, next_state, done)
            frame = 0
        else:
            frame += 1
        
        
        while not info[0] and not done:
            next_state, reward, done, info = env.step(4)
            
        state = next_state

        
        if done:
            print("episode: {}/{}, score: {}, time: {}, epsilon:{:.2}".format(e,n_episodes,env.last_score,time, agent.epsilon))
            break
            
    if len(agent.memory) > batch_size:
        agent.replay(batch_size)
    
    
    if e % 1000 == 0 and e != 0:
        agent.save(output_dir+ "/" + name + "_{:05d}.hdf5".format(e + 1000))
    
        
#agent.save(output_dir+ "/" + name + "_{:05d}.hdf5".format(n_episodes-1))
#agent.save(output_dir+ "/" + name + "_02000.hdf5")
```
